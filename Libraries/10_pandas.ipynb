
## üìä Creating Series and DataFrames

### üß† Theory

- `Series`: A one-dimensional labeled array capable of holding any data type.
- `DataFrame`: A two-dimensional size-mutable, potentially heterogeneous tabular data.

### üõ†Ô∏è Usage

- `pd.Series(data, index=index)`
- `pd.DataFrame(data, index=index, columns=columns)`


```python

# Create a Series
series = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print("Series:\n", series)

# Create a DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
})
print("\nDataFrame:\n", df)
```


## üìã DataFrame Attributes

### üß† Theory

Common attributes used to inspect the DataFrame structure:

| Attribute      | Description                     |
|----------------|---------------------------------|
| `df.shape`     | Tuple of (rows, columns)        |
| `df.size`      | Total number of elements        |
| `df.columns`   | Column labels                   |
| `df.index`     | Row index                       |
| `df.dtypes`    | Data type of each column        |
| `df.values`    | Numpy representation of values  |


```python

print("Shape:", df.shape)
print("Size:", df.size)
print("Columns:", df.columns)
print("Index:", df.index)
print("Data Types:\n", df.dtypes)
print("Values:\n", df.values)
```


## üîç Indexing and Selection

### üß† Theory

Access elements using:
- `df['col_name']` for a column
- `df.loc[]` for label-based selection
- `df.iloc[]` for integer-based selection

###  Examples


```python

# Select column
print(df['Name'])

# Select row by label
print(df.loc[0])

# Select row by index position
print(df.iloc[1])
```

## Value_counts()

It iterates through all the values in your Series and identifies every distinct (unique) value. For each unique value, it counts how many times it appears in the Series.. It gives you an immediate understanding of the distribution of values in a column. For example, if you have a column for "Gender," value_counts() would quickly tell you how many "Male" and "Female" entries there are

```python
# value_counts() function is used on categorical data to find frequency

data['winner'].value_counts()
```

## Drop Duplicates

It scans the DataFrame (or a specific set of columns) to find rows that are exact copies of other rows. Once identified, it removes these duplicate rows, leaving only unique rows.

```python
data.drop_duplicates(subset=["city",'season']).shape
```

```python
data.head()
```

```python
# Title winner Season wise

data.drop_duplicates(subset='season', keep='last')[["season",'winner']].sort_values('season')
```

## Project 2

```python
data2 = pd.read_csv("deliveries.csv")
```

```python
data2.shape
```

```python
data2.head()
```

### Batsman with most number of runs

```python
season2 = data2.groupby('batsman')
```

```python
season2['batsman_runs'].sum().sort_values(ascending=False).head(6)
```

### Batsman with most number of 6s

```python
only6 = (data2[data2['batsman_runs']==6]).groupby('batsman')
```

```python
only6["batsman_runs"].value_counts().sort_values(ascending=False).head(5)
```

### 3 Teams againts which DA Warner scored most runs 

```python
mostruns = (data2[data2['batsman'] == 'DA Warner']).groupby('bowling_team')
mostruns['batsman_runs'].sum().sort_values(ascending=False)
```

### Function for reproducibility

```python
def runs_scored(name):

    mostruns = (data2[data2['batsman'] == name]).groupby('bowling_team')
    runs = mostruns['batsman_runs'].sum().sort_values(ascending=False)
    return runs

runs_scored("V Kohli")
```
