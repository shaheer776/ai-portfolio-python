# Getting Data (Titanic dataset)

```python
import pandas as pd

df = pd.read_csv('dataset.csv') # To read any CSV file
df.head()
```

# Univariate Analysis

## 1. Categorical Data

### a. Countplot

```python
import seaborn as sns

sns.countplot(df, x="Survived")
df['Survived'].value_counts()
```

```python
sns.countplot(df, x='Pclass')
```

```python
sns.countplot(df, x='Sex')
```

### b. Pie chart

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')  # Shows percentage of people who could and couldn't survived
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')  # Shows percentage of people in various classes
```

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%.2f')  # Shows percentage of gender of people aboard titanic
```

## 2. Numerical Data

### a. Histogram

```python
import matplotlib.pyplot as plt

plt.hist([df['Age']]) # Histogram shows the distribution of data
```

### b. Distplot

```python
sns.distplot(df['Age']) # The line is the Probability density function
```

### c. Boxplot

```python
sns.boxplot(df, x='Age')
```

```python
df['Age'].min()
```

```python
df['Age'].max()
```
