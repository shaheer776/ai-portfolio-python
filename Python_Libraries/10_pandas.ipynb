
## üìä Creating Series and DataFrames

### üß† Theory

- `Series`: A one-dimensional labeled array capable of holding any data type.
- `DataFrame`: A two-dimensional size-mutable, potentially heterogeneous tabular data.

### üõ†Ô∏è Usage

- `pd.Series(data, index=index)`
- `pd.DataFrame(data, index=index, columns=columns)`


```python

# Create a Series
series = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print("Series:\n", series)

# Create a DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
})
print("\nDataFrame:\n", df)
```

### isin()

It takes a Series or DataFrame and a collection of values (like a list, tuple, Series, or set) as input. Returns a boolean Series/DataFrame: For each element in the original Series/DataFrame, it checks if that element exists within the provided collection of values.

- If the element is found in the collection, it returns True.
- If the element is not found, it returns False.

## Pivot Table

It transforms data from a list-like format into a table with a new structure.

Key Parameters:

    data: The DataFrame from which to create the pivot table.
    values: Column(s) to aggregate. If not specified, all numeric columns will be aggregated.
    index: Column(s) to use for the new DataFrame's index (rows). Can be a single column name or a list of column names.
    columns: Column(s) to use for the new DataFrame's columns. Can be a single column name or a list of column names.
    aggfunc: Aggregation function(s) to use. Can be a single function (e.g., 'mean', 'sum', np.mean), a list of functions, or a dictionary mapping columns to functions.
        Common strings: 'mean', 'sum', 'count', 'min', 'max', 'median', 'first', 'last'.
        NumPy functions: np.mean, np.sum, np.median, etc.
    fill_value: Value to replace missing values (NaN) in the pivot table output. Useful for making the table look cleaner or for further calculations.
    margins: Boolean (default False). If True, adds row and column subtotals/totals (called "margins").
    dropna: Boolean (default True). If True, don't include columns whose entries are all NaN.
    margins_name: String (default 'All'). Name of the row/column that will contain the totals when margins=True.

### Correlation

Correlation describes the strength and direction of a linear relationship between two or more variables. It tells you how closely changes in one variable are associated with changes in another.
- Positive Correlation: As one variable increases, the other variable also tends to increase. (e.g., more hours studied, higher test scores). The correlation coefficient will be positive (between 0 and +1).
- Negative Correlation: As one variable increases, the other variable tends to decrease. (e.g., more hours exercising, lower weight). The correlation coefficient will be negative (between 0 and -1).
- No Correlation (Zero Correlation): There is no consistent linear relationship between the variables.

## Missing Values

### dropna()
dropna()  function is used to remove missing values (NaN, None, NaT, etc.) from a Series or DataFrame.


Key Parameters:

    axis: Determines whether to drop rows or columns.
        0 or 'index' (default): Drops rows containing missing values.
        1 or 'columns': Drops columns containing missing values.
    how: Specifies how many missing values are needed to trigger a drop.
        'any' (default): Drops the row/column if any missing values are present.
        'all': Drops the row/column only if all values are missing (i.e., the entire row/column is empty).
    thresh: Requires a minimum number of non-NA values for the row/column to be kept. If thresh=3, a row will only be kept if it has at least 3 non-missing values. This is useful for keeping rows that are mostly complete.
    subset: Column label or list of column labels. When dropping rows (axis=0), subset specifies which columns to consider for missing values. Only rows with missing values in these specific columns will be dropped.
    inplace: Boolean (default False). If True, the operation is performed on the original DataFrame/Series and returns None. If False, a new object is returned

### fillna()

fillna() function is used to fill (impute) missing values (NaN, None, NaT, etc.) with a specified value or method


Key Parameters:

    value: The scalar value (e.g., 0, '-', mean_of_column) to use for filling missing values. This is the simplest way.
    method: A more sophisticated way to fill missing values, often based on adjacent values.
        'pad' or 'ffill' (forward fill): Fills NaN values with the last valid observation. Useful for time series.
        'bfill' or 'backfill' (backward fill): Fills NaN values with the next valid observation. Useful for time series.
    axis: Determines whether to fill along rows (0) or columns (1).
        0 or 'index' (default): Fills missing values down each column.
        1 or 'columns': Fills missing values across each row.
    limit: For method fills, the maximum number of consecutive NaN values to fill.
    inplace: Boolean (default False). If True, the operation is performed on the original DataFrame/Series and returns None. If False, a new object is returned.
